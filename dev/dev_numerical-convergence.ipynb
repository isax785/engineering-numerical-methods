## Functions with Arguments

Function types:
-  `y = f(x)`, i.e. single input - single output;
-  `y=f(x, x1, x2, x3, ...)`, i.e. multiple inputs - single output.

```python
# Monotonic Incrementing - Single Input - trend is NOT declared
settings = {'tol': 0.005,
            'delta': 5,
            'delta_scaler': 2,
            'x_0': 100,
            'x_min': None,
            'x_max': None,
            'y_t': 4.5,
            'printout': True}

def function(x:float)->float:
    return x**2

x, y = nc_function_args(settings, function)
```

```python
# Monotonic Decrementing - Single Input
settings = {'tol': 0.005,
            'delta': 5,
            'delta_scaler': 2,
            'x_0': 100,
            'x_min': None,
            'x_max': None,
            'y_t': - 50,
            'trend': False,
            'printout': True}

def function(x:float)->float:
    return - x**2

x, y = nc_function_args(settings, function)
```

```python
# Monotonic Incrementing - Multiple Inputs
settings = {'tol': 0.005,
            'delta': 5,
            'delta_scaler': 2,
            'x_0': 100,
            'x_min': None,
            'x_max': None,
            'y_t': 50,
            'trend': True,
            'printout': True}

def function(x:float, x1:float, x2:float)->float:
    return x**2 + x1 + 3 * x2

x1x2 = (2, 3)

x, y = nc_function_args(settings, function, *x1x2)
```

```python
# Monotonic Decrementing - Multiple Inputs
settings = {'tol': 0.005,
            'delta': 5,
            'delta_scaler': 2,
            'x_0': 100,
            'x_min': None,
            'x_max': None,
            'y_t': - 500,
            'trend': False,
            'printout': True}

def function(x:float, x1:float, x2:float)->float:
    return - x**2 - x1 - 3 * x2

x1x2 = (2, 3)

x, y = nc_function_args(settings, function, *x1x2)
```

## Functions with Dictionaries

```python
# Monotonic Incrementing - Multiple Inputs - trend is NOT declared
settings = {'tol': 0.005,
            'delta': 5,
            'delta_scaler': 2,
            'x_name': 'x',
            'x_0': 100,
            'x_min': None,
            'x_max': None,
            'y_t_name': 'sum',
            'y_t': 50,
            'trend': True,
            'printout': True}

def function_dict(inp_dict:dict)->dict:
    x = inp_dict.get('x')
    z = inp_dict.get('z', 0) 
    res = {'sum': x - z,
           'prod': x * z}
    return res

inp_dict = {'x': 12,
            'z': 3}

inp_dict, res = nc_function_dict(settings, function_dict, inp_dict)
```
